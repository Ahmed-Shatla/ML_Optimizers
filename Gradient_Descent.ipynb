```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
   return x**2 - 4*x + 6

def df(x):
   return 2*x - 4
```

```python
def gradient_descent(init_x, lr, itr):
  '''
  this is a gradient descent function
  Args:
      init_x (int): initial X
      lr (float): learning rate
      itr (int): number of iterations

  Returns:
      local minimum, updating history
  '''
  x = init_x
  x_history=[x]

  for i in range(itr):
    gradient = df(x)
    x = x - lr*gradient
    x_history.append(x)

  return x,x_history
```

```python
initial_x = 0
learning_rate = 0.1
num_iterations = 30

x, x_history = gradient_descent(initial_x, learning_rate, num_iterations)

print("Local minimum: {:.2f}".format(x))
```

```python
x_vals = np.linspace(-1, 4, 100)
plt.plot(x_vals, func(x_vals))
plt.plot(x_history, func(np.array(x_history)), 'rx')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
```
